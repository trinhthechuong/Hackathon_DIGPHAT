```python
import pandas as pd
import re
import os

# File path
file_path = "/Users/thechuongtrinh/Workspace/Hackathon_DIGPHAT/Data/20260211_TurboID_v5-v6_MS_data_modified.xlsx"

# Load the dataset
df = pd.read_excel(file_path)

def clean_subcellular_location(text):
    if not isinstance(text, str):
        return text
    
    # 1. Remove the 'SUBCELLULAR LOCATION:' prefix
    clean_text = re.sub(r'^SUBCELLULAR LOCATION:\s*', '', text, flags=re.IGNORECASE)
    
    # 2. Remove ECO citations and PubMed IDs like {ECO:0000269|PubMed:16128801}
    clean_text = re.sub(r'\{.*?\}', '', clean_text)
    
    # 3. Split by '.' and ';' to separate individual locations
    parts = re.split(r'[.;]', clean_text)
    
    # 4. Clean up whitespace and filter out 'Note=' sections or empty strings
    locations = []
    for part in parts:
        part = part.strip()
        if part and not part.lower().startswith('note='):
            locations.append(part)
            
    return ", ".join(locations)

# Apply the cleaning function to the 'Unnamed: 19' column
column_target = 'Unnamed: 19'

if column_target in df.columns:
    # Create a new column or overwrite the existing one
    df['Cleaned Location'] = df[column_target].apply(clean_subcellular_location)
    print(f"Successfully processed {column_target}")
    
    # Display the first few rows to verify
    display(df[[column_target, 'Cleaned Location']].head(10))

    # Save the cleaned dataframe to a new Excel file
    output_path = file_path.replace(".xlsx", "_processed.xlsx")
    df.to_excel(output_path, index=False)
    print(f"\nSaved processed data to: {output_path}")

else:
    print(f"Column '{column_target}' not found. Available columns are:")
    print(df.columns.tolist())
```
